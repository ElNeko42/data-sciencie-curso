### Ejercicio 1:

Usando un mapa de **`folium`** representa alguna información de tu interés, esta información debe tener los datos de ubicación (latitud y longitud) para poder ser representados en el map, algunos ejemplos pueden ser:

- La ubicación de museos en Madrid.
- Lugar de eventos culturales de tu ciudad.
- Lista de restaurantes más famosos de España.
- Capital de los paises de Europa.
- La ubicación de los estadios donde se jugarán los partidos del Mundial Qatar 2022.

```python
# Crear un mapa centrado en Madrid
mapa_madrid = folium.Map(location=[40.4167, -3.70325], zoom_start=12)

# Lista de museos y sus coordenadas (ejemplo con datos ficticios)
museos = [
    {"nombre": "Museo del Prado", "coordenadas": [40.413722, -3.692412]},
    {"nombre": "Museo Reina Sofía", "coordenadas": [40.408737, -3.693822]},
    {"nombre": "Museo Thyssen-Bornemisza", "coordenadas": [40.416064, -3.694676]},
    # Añade más museos aquí
]

# Añadir un marcador para cada museo
for museo in museos:
    folium.Marker(
        location=museo["coordenadas"],
        popup=museo["nombre"],
        icon=folium.Icon(color='red', icon='info-sign')
    ).add_to(mapa_madrid)


mapa_madrid

```

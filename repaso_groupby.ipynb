# Groupby

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("mod3-data-science/Data/titanic.csv")
```

```python
df[["PassengerId", "Age", "Fare"]].agg(lambda x: x**2, axis = 0)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df["Fare"])
plt.show()

sns.histplot(df["Fare"].apply(lambda x: np.log(x)))
plt.show()
```

```python
df[["PassengerId", "Age", "Fare"]].apply(lambda x: x**2)
```

```python
df.groupby("Sex")[["Fare"]].median(numeric_only=True)
```

```python
df[df["Sex"] == "male"]["Fare"].median()
```

```python
df[df["Sex"] == "female"]["Fare"].median()
```

```python
df.groupby("Pclass").mean(numeric_only=True)#[["Survived", "Fare"]]
```
